```python
def impute_age(cols):
    age=cols[0]
    education=cols[1]
    city=cols[2]
    if(pd.isnull(age)):
        if(education=='Matriculation'):
            if(city==1):
                return 42
            elif(city==2):
                return 40
            elif(city==3):
                return 37
            elif(city==4):
                return 33
        elif(education=='High School Diploma'):
            if(city==1):
                return 32
            elif(city==2):
                return 32
            elif(city==3):
                return 31
            elif(city==4):
                return 30
        elif(education=='Bachelors'):
            if(city==1):
                return 43
            elif(city==2):
                return 43
            elif(city==3):
                return 43
            elif(city==4):
                return 43
        elif(education=='Masters'):
            if(city==1):
                return 45
            elif(city==2):
                return 45
            elif(city==3):
                return 46
            elif(city==4):
                return age
        elif(education=='No Qualification'):
            if(city==1):
                return 27
            elif(city==2):
                return 30
            elif(city==3):
                return 29
            elif(city==4):
                return 36.5
    else:
        return age
```

```python
df1['age'] = df1[['age','education','city_tier']].apply(impute_age,axis=1)
```

```python
df.head()
```

```python
df=df.replace({'difficulty_level': {'easy' :0,'intermediate' :1,'hard':2,'vary hard' :3},
                     'education' :{'No Qualification' :0,'Matriculation' :1,'High School Diploma' :2,'Bachelors' :3,'Masters':4}, 
                    'program_type':{'S':0,'T':1,'U':2,'V':3,'W':4,'X':5,'Y':6,'Z':7},
                    'test_type':{'offline':0,'online':1},
                    'gender':{'M':1,'F':0}, 
                    'is_handicapped':{'N':0,'Y':1} 
                    })
df.head()
```

```python
df.drop(['id','age'],axis=1,inplace=True)
df.head()
```

```python
train=df[:73147]
test=df[73147:]
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5, weights="uniform")
col=['trainee_id','gender','education','city_tier', 'trainee_engagement_rating'] 
X_tr=train[col]
X_test=test[col] 
X_tr=imputer.fit_transform(X_tr)
X_test=imputer.transform(X_test) 
train=train.join(pd.DataFrame({'imp':X_tr[:,- 1]}))
test=test.join(pd.DataFrame({'imp':X_test[:,- 1]}))

train=train.drop(['trainee_engagement_rating'],axis=1)
test=test.drop(['trainee_engagement_rating'],axis=1)

Id=train.program_id.unique()
for i in range(len(Id)):
    train=train.replace({Id[i]:i})
    test=test.replace({Id[i]:i})
```

```python
from catboost import CatBoostClassifier,Pool
from sklearn.model_selection import train_test_split
```

```python
train_y=dependent_var
x_train,x_valid,y_train,y_valid=train_test_split(train,train_y,train_size =0.8)
```

```python
cat_features=[0,1,3,6] 
#these are the indices of "program_id", "program_type",
#"test_id", "trainee_id" 
```

```python
train_pool=Pool(train, train_y, cat_features=cat_features)
test_pool=Pool(test, cat_features =cat_features) 
```

```python
weight=[1, 0.43]
```

```python
model = CatBoostClassifier(
        iterations=5000,logging_level="Silent",early_stopping_rounds=500,
    use_best_model=True,custom_loss=["AUC"], class_weights=weight, 
        eval_metric="AUC")
```

```python
model.fit(  x_train,y_train, 
        cat_features=cat_features,
        eval_set=(x_valid,y_valid),
        plot=True 
        )
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_valid, model.predict_proba(x_valid) [:,1]) 
```

```python
feature_importances=model.get_feature_importance(train_pool) 
feature_names=train.columns
for score, name in sorted(zip(feature_importances, feature_names), reverse=True):
    print('{}: {}'.format(name, score))
```
